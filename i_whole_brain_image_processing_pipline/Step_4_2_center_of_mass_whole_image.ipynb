```python
import tifffile
```

```python
img = tifffile.imread('./Step5/downsampled_image_z4_y2_x2.tiff')
print(img.shape)
```

```python
from scipy.ndimage import center_of_mass
import numpy as np

labels = np.unique(img)
labels = labels[labels != 0]
coms = center_of_mass(img > 0, labels=img, index=labels)
results = list(zip(labels, coms))
```

```python
import csv

csv_path = "./Step5/center_of_mass_results.csv"

with open(csv_path, mode='w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Label', 'Z', 'Y', 'X'])  # Header
    for label, (z, y, x) in results:
        writer.writerow([label, z, y, x])
```
